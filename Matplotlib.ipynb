# Line Plot

```python
x = [1,2,3,4,5]
y = [2,3,4,7,10]

plt.plot(x,y)
plt.xlabel('X values')
plt.ylabel('Y values')
plt.title('line x vs y')
plt.show()
```

```python
x,y
```

# Bar PLot

```python
plt.bar(x,y,edgecolor='black',color='white',label='bar explaining evertything')
plt.plot(x,y,color='red',label='line explaining some stuff')
plt.xlabel('X values')
plt.ylabel('Y values')
plt.title('line x vs y')
plt.legend()
plt.xlim(2,6)
plt.grid(True)
plt.show()
```

# Subplot

```python
data = np.random.randn(1000)
plt.subplot(2,2,1)
plt.hist(data,bins=40,edgecolor='black')

plt.subplot(2,2,4)
sns.histplot(data,bins=40,kde=True)
plt.show()

plt.savefig('myimg.png') # To save the graph on local
```

```python
x = np.linspace(0,10,100)
y1 = np.sin(x)
y2= np.cos(x)
```

```python

plt.plot(x, y1, label='Sin')
plt.plot(x, y2, label='Cos')
plt.xlabel('X Axis Label')
plt.ylabel('Y Axis Label')
plt.title('Plot Title')
plt.xlim(0, 5)  # Set x-axis limits
plt.ylim(-1,1)  # Set y-axis limits
plt.xticks(np.arange(0, 10, 0.5))  # Set custom tick locations on x-axis
plt.yticks([-1, 0, 1], ['Min', 'Mid', 'Max'])  # Set custom tick labels on y-axis
plt.grid(True)  # Show grid lines
plt.legend(loc=5)  # Show legend
plt.tight_layout()  # Adjust layout
# plt.savefig('plot.png')  # Save plot to file
plt.show()
```
